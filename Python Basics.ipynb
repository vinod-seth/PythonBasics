# Overview of OOP Terminology

 -  <b>Class</b> − A user-defined blueprint for an object that defines a set of attributes that characterize any object of the class. The attributes are data members (class variables and instance variables) and methods, accessed via dot notation.
 -  <b>Class variable</b> − A variable that is shared by all instances of a class. Class variables are defined within a class but outside any of the class's methods. Class variables are not used as frequently as instance variables are.

 -  <b>Data member</b> − A class variable or instance variable that holds data associated with a class and its objects.

-  <b>Function overloading</b> − The assignment of more than one behavior to a particular function. The operation performed varies by the types of objects or arguments involved.

-  <b>Instance variable</b> − A variable that is defined inside a method and belongs only to the current instance of a class.

-  <b>Inheritance</b> − The transfer of the characteristics of a class to other classes that are derived from it.

-  <b>Instance</b> − An individual object of a certain class. An object obj that belongs to a class Circle, for example, is an instance of the class Circle.

-  <b>Instantiation</b> − The creation of an instance of a class.

-  <b>Method</b> − A special kind of function that is defined in a class definition.

-  <b>Object</b> − A unique instance of a data structure that's defined by its class. An object comprises both data members (class variables and instance variables) and methods.

-  <b>Operator overloading</b> − The assignment of more than one function to a particular operator.

# Inheritance

Inheritance is a way of creating a new class for using details of an existing class without modifying it. The newly formed class is a derived class (or child class). Similarly, the existing class is a base class (or parent class).

```python
# parent class
class Bird:
    
    def __init__(self):
        self.a = 10;
        print("Bird is ready")

    def whoisThis(self):
        print("Bird")
    def hello(self):
        print("hello")
    def swim(self):
        print("Swim faster %d" % self.a)
```

```python
birdy = Bird()
birdy.whoisThis()
```

```python
# child class
# Penguin is a bird
# penguin is derived from Bird
class Penguin(Bird): # The child class inherits the functions of parent class

    def __init__(self):
        # call super() function
        super().__init__()   #This allows us to run the __init__() method of the parent class inside the child class.
        print("Penguin is ready")

    def whoisThis(self): # overriding the whoisThis behaviour
        print("Penguin")

    def run(self):
        print("Run faster")

#swim()
# not whoisthis. because penguin is overriding the whoisthis method
```

```python
peggy = Penguin()
```

```python
peggy.whoisThis()
peggy.swim()
```

#  Encapsulation

we can <b><i><font size="3">restrict access to methods and variables</font></i></b>. This prevents data from direct modification which is called encapsulation. In Python, we denote private attributes using underscore as the prefix i.e single<b> _ or double \__</b>

```python
class Computer:
  
    def __init__(self):
        # __maxprice is a private data member
        self.__maxprice = 900  # private variable
        
    # below is the declaration of a private function
    def __setMaxpriceP(self, price): # privtae can't be invoked outside of a class
        self.__maxprice = price
        
    def sell(self):
        print("Selling Price: {}".format(self.__maxprice))

    def setMaxPrice(self, price):
        self.__maxprice = price
    def setUsingPrivateFn(self,price): # public function which can be invoked using object
        self.__setMaxpriceP(price)

c = Computer()
c.sell()

# change the price
c.__maxprice = 1000 # We tried to modify the price. However, 
                    # we can't change it because Python treats the __maxprice as private attributes.
c.sell()

# using setter function
c.setMaxPrice(1000)
c.sell()
```

```python
# __ it means private method. it is invoked using instance or object. beecause we are trying to acces the private method
# outside the class
c.__setMaxPriceP(1000)
```

```python
# private members can not be accessed outside the class
# we define public methods to access private data
c.setUsingPrivateFn(2000)
c.sell()
```

- <b>Instance Methods</b>
Instance methods are functions that are defined inside a class and can only be called from an instance of that class. Just like .\__init__(), an instance method’s first parameter is always self.

```python
class Dog:
    species = "Canis familiaris"  # class data member

    def __init__(self, name, age):  # instance method
        self.name = name
        self.age = age
    def __str__(self):              
        return f" i am {self.species}"
    # Instance method
    def description(self):
        return f"{self.name} is {self.age} years old"

    # Another instance method
    def speak(self, sound):
        return f"{self.name} says {sound}"
    # f is for format   
```

This Dog class has two instance methods:

- <b>description()</b>:-  returns a string displaying the name and age of the dog.
- <b>dspeak()</b>:-  has one parameter called sound and returns a string containing the dog’s name and the sound the dog makes.

# Abstract Base Classes in Python (abc)

An abstract method is a method that is declared, but contains no implementation. Abstract classes can not be instantiated, and its abstract methods must be implemented by its subclasses.

```python
import abc
# abc is python module for abstract methods
class Shape(metaclass=abc.ABCMeta):
   counter = 0
   @abc.abstractmethod
   def area(self):
      pass
#    @abc.abstractmethod
#    def abc(self):
class Rectangle(Shape):
   def __init__(self, x,y):
      self.l = x
      self.b = y
   def area(self):
      return self.l*self.b
```

# Class decorators

- <b>@classmethod</b>:- this decorator declares a method as class level method. all methods decorated with @classmethod decorator accepts a <b><font size = 3>cls</font></b> parameter instaed of <b><font size = 3>self</font></b>

- <b>@staticmethod</b>:- A static method doesn’t have the calling object passed into it as the first parameter. 
    it doesn’t have access to the rest of the class or instance at all, We can call them from an instance or a class object

```python
class Person:
    TITLES = ('Dr', 'Mr', 'Mrs', 'Ms')

    def __init__(self, name, surname):
        self.name = name
        self.surname = surname

    def fullname(self): # instance method
        # instance object accessible through self
        return "%s %s" % (self.name, self.surname)

    @classmethod
    def allowed_titles_starting_with(cls, startswith): # class method
        # class or instance object accessible through cls
        return [t for t in cls.TITLES if t.startswith(startswith)]
    # if t.startswith(startswith)
    #   t.append[t] 
    # t = Dr
    # if t.startswith("M"):
    #   t.append[t]   false

    @staticmethod
    def allowed_titles_ending_with(endswith): # static method
        # no parameter for class or instance object
        # we have to use Person directly
        
        #print(cls.TITLES)
        #print(self.name)
        return [t for t in Person.TITLES if t.endswith(endswith)]


jane = Person("Jane", "Smith")
```

- <b>setter and getter for property</b>:- There are also decorators which we can use to define a setter and a deleter for our attribute (a deleter will delete the attribute from our object). The getter, setter and deleter methods must all have the same name:

```python
class Person:
    def __init__(self, name, surname):
        self.name = name
        self.surname = surname

    @property
    def fullname(self):
        return "%s %s" % (self.name, self.surname)

    @fullname.setter
    def fullname(self, value):
        # this is much more complicated in real life
        print('setter is called')
        # "Jane Doe"
        name, surname = value.split(" ", 1)
        self.name = name
        self.surname = surname

    @fullname.deleter
    def fullname(self):
        print('deleter is called')
        del self.name
        del self.surname

```

# Multiple inheritance

```python
class Person:
    def __init__(self, name, surname, number):
        print("constrcutor")
        self.name = name
        self.surname = surname
        self.number = number


class LearnerMixin:
    def __init__(self):
        print("constructor invoked")
        self.classes = []

    def enrol(self, course):
        self.classes.append(course)


class TeacherMixin:
    def __init__(self):
        self.courses_taught = []

    def assign_teaching(self, course):
        self.courses_taught.append(course)

# * args is tuple  ()
# ** kwargs is dictionary
class Tutor(Person, LearnerMixin, TeacherMixin):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        super(Person, self).__init__(LearnerMixin)

jane = Tutor("Jane", "Smith", "SMTJNX045")
#jane.enrol("a_postgrad_course")
#jane.assign_teaching("an_undergrad_course")
```

```python
# tuple:- it does not allow duplicate values
# using * args
def adder(*num):  #  num = (3,5) is a list/tuple
    sum = 0
    
    for n in num:
        #1. n = 3
           #sum = 3 + 0 = 3
        #2. n =5
           # sum = 3 + 5 = 8
        sum = sum + n

    print("Sum:",sum)

adder(1,2,3,4)
```
