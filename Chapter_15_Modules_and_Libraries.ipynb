# Chapter 15: Modules & Libraries

## What are Modules?

A **module** is a Python source file that you load into your program to make its contents available — functions, classes, constants, and more. A **library** (or package) groups several related modules together under a single name, giving you a larger, organized toolkit to draw from.

---

## Importing Modules

```python
# Import entire module
import math

# Use module functions
print(math.pi)           # Value of pi
print(math.sqrt(16))     # Square root
print(math.floor(3.7))   # Round down
print(math.ceil(3.2))    # Round up
```

```python
# Import specific functions
from math import sqrt, pi, ceil

print(pi)
print(sqrt(25))
print(ceil(4.1))
```

```python
# Import with alias (short name)
import math as m

print(m.pi)
print(m.sqrt(36))
```

## The random Module

```python
import random

# Random integer
num = random.randint(1, 10)
print(f"Random number 1-10: {num}")

# Random float
float_num = random.random()  # Between 0 and 1
print(f"Random float: {float_num}")

# Random choice from list
colors = ["red", "blue", "green"]
choice = random.choice(colors)
print(f"Random color: {choice}")
```

```python
# Shuffle list
cards = ["♠", "♥", "♦", "♣"]
random.shuffle(cards)
print("Shuffled:", cards)
```

## The datetime Module

```python
from datetime import datetime, timedelta

# Current date and time
now = datetime.now()
print(f"Current date and time: {now}")
print(f"Year: {now.year}")
print(f"Month: {now.month}")
print(f"Day: {now.day}")
```

```python
# Create specific date
birthday = datetime(2000, 12, 25)
print(f"Birthday: {birthday}")

# Calculate age
age = (datetime.now() - birthday).days // 365
print(f"Age in years: {age}")
```

```python
# Add days
today = datetime.now()
next_week = today + timedelta(days=7)
print(f"Today: {today.date()}")
print(f"Next week: {next_week.date()}")
```

## The os Module

```python
import os

# Get current directory
print(f"Current directory: {os.getcwd()}")

# List files in directory
files = os.listdir(".")
print(f"\nFirst 5 items in directory:")
for file in files[:5]:
    print(f"  - {file}")
```

```python
# Check if file exists
if os.path.exists("my_file.txt"):
    print("File exists!")
else:
    print("File does not exist")

# Check if it's a file or directory
if os.path.isfile("my_file.txt"):
    print("It's a file")
elif os.path.isdir("my_file.txt"):
    print("It's a directory")
```

## List Comprehension (Bonus!)

```python
# Create lists in one line

# Squares of 1-5
squares = [x**2 for x in range(1, 6)]
print(f"Squares: {squares}")

# Even numbers 1-10
evens = [x for x in range(1, 11) if x % 2 == 0]
print(f"Evens: {evens}")

# Uppercase letters
words = ["python", "java", "javascript"]
upper_words = [w.upper() for w in words]
print(f"Uppercase: {upper_words}")
```

## Popular Libraries (Beyond Basics)

Once you are comfortable with the language fundamentals, these widely-used libraries are worth exploring:

- **NumPy**: Provides fast multi-dimensional arrays and a broad set of mathematical operations — the foundation for most scientific computing in Python.
- **Pandas**: Introduces the DataFrame, a table-like structure built for loading, cleaning, reshaping, and summarising real-world datasets.
- **Matplotlib**: A plotting library that turns raw numbers into charts, graphs, and figures you can inspect and share.
- **Requests**: Simplifies sending HTTP requests so you can interact with web APIs and download remote resources in just a few lines.
- **Django / Flask**: Web frameworks for building server-side applications — Django is full-featured and opinionated; Flask is lightweight and flexible.
- **Pygame**: A set of modules for writing games and multimedia applications, handling graphics, sound, and user input.
- **TensorFlow / PyTorch**: Deep learning frameworks used to design, train, and deploy neural network models at scale.
