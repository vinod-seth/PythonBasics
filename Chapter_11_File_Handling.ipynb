# Chapter 11: File Handling

## Working with Files

Python can **read**, **write**, and **modify** files

---

## Creating and Writing Files

```python
# Create and write to a file
file = open("my_file.txt", "w")  # "w" = write mode
file.write("Hello, World!\n")
file.write("I am learning Python.\n")
file.write("File handling is fun!")
file.close()  # Always close!

print("File created successfully!")
```

**File Modes:**
- `"r"` - Read (file must exist)
- `"w"` - Write (creates new or overwrites)
- `"a"` - Append (adds to end)
- `"x"` - Create (fails if exists)

---

## Reading Files

```python
# Read entire file
file = open("my_file.txt", "r")
content = file.read()
file.close()

print(content)
```

```python
# Read line by line
file = open("my_file.txt", "r")
line1 = file.readline()
line2 = file.readline()
line3 = file.readline()
file.close()

print(repr(line1))
print(repr(line2))
print(repr(line3))
```

```python
# Read all lines as list
file = open("my_file.txt", "r")
lines = file.readlines()
file.close()

for line in lines:
    print(repr(line))
```

---

## Using with Statement (Best Practice)

```python
# Using 'with' - automatically closes file
with open("my_file.txt", "r") as file:
    content = file.read()
    print(content)

# File is automatically closed here
```

---

## Appending to Files

```python
# Append to file (add at the end)
with open("my_file.txt", "a") as file:
    file.write("\nAdded this line!")

# Read and verify
with open("my_file.txt", "r") as file:
    print(file.read())
```

---

## Working with Lines

```python
# Write multiple lines
with open("names.txt", "w") as file:
    file.write("Alice\n")
    file.write("Bob\n")
    file.write("Charlie\n")

# Read and process
with open("names.txt", "r") as file:
    for line in file:
        name = line.strip()  # Remove \n
        print(f"Hello, {name}!")
```

---

## Working with CSV Files

```python
# Create CSV file
with open("students.csv", "w") as file:
    file.write("Name,Age,Grade\n")
    file.write("Alice,20,A\n")
    file.write("Bob,19,B\n")
    file.write("Charlie,21,A\n")

# Read and parse CSV
with open("students.csv", "r") as file:
    for line in file:
        data = line.strip().split(",")
        print(f"Name: {data[0]}, Age: {data[1]}, Grade: {data[2]}")
```

---

## Practical Example: Count Words

```python
# Create a text file
with open("sample.txt", "w") as file:
    file.write("Python is awesome. Python is fun. Python is powerful.")

# Count specific word
with open("sample.txt", "r") as file:
    content = file.read()
    count = content.count("Python")
    print(f"'Python' appears {count} times")
```

## Key Takeaways

 Open files: `open(filename, mode)`
 Modes: "r"(read), "w"(write), "a"(append)
 Read: `read()`, `readline()`, `readlines()`
 Write: `write()`
 Always close files or use `with` statement
 Process files line by line when large
