# Chapter 10: String Methods & Formatting

## What are String Methods?

Strings in Python come with a built-in toolkit of operations you can call directly on any string value. The syntax follows the pattern `string.method()`, where the method acts on the string that precedes the dot.

---

## Case Methods

```python
text = "Hello World"

# Convert to uppercase
print(text.upper())

# Convert to lowercase
print(text.lower())

# Capitalize first letter
print(text.capitalize())

# Swap case
print(text.swapcase())
```

---

## Finding & Replacing

```python
text = "I love Python programming"

# Find position (returns index, -1 if not found)
print("Position of 'Python':", text.find("Python"))
print("Position of 'Java':", text.find("Java"))

# Check if text contains
print("'Python' in text?", "Python" in text)
```

```python
# Replace text
text = "I love Java. Java is fun"
new_text = text.replace("Java", "Python")
print(new_text)

# Replace only first occurrence
new_text = text.replace("Java", "Python", 1)
print(new_text)
```

---

## Splitting & Joining

```python
# Split string into list
sentence = "Python is amazing and fun"
words = sentence.split()
print(words)

# Split by specific delimiter
csv = "apple,banana,orange"
fruits = csv.split(",")
print(fruits)
```

```python
# Join list into string
words = ["Python", "is", "awesome"]
sentence = " ".join(words)
print(sentence)

# Join with different separator
items = ["apple", "banana", "orange"]
csv = ",".join(items)
print(csv)
```

---

## Trimming Strings

```python
# Remove leading & trailing spaces
text = "  Hello World  "
print("Original:", repr(text))
print("Trimmed:", repr(text.strip()))

# Remove only left side
print("Left trimmed:", repr(text.lstrip()))

# Remove only right side
print("Right trimmed:", repr(text.rstrip()))
```

---

## Checking String Properties

```python
# Check if all uppercase
print("HELLO".isupper())
print("Hello".isupper())

# Check if all lowercase
print("hello".islower())
print("Hello".islower())

# Check if all digits
print("12345".isdigit())
print("123a5".isdigit())

# Check if all alphabetic
print("hello".isalpha())
print("hello123".isalpha())
```

## String Formatting

Embedding variable values inside a string is one of the most common tasks in any Python program — whether you are displaying results to a user, building a log message, or constructing a data record. Python offers three distinct approaches:

Three ways to format strings:

### Method 1: f-strings (Recommended)

```python
name = "Alice"
age = 25
score = 95.5

# f-string - put variables inside {}
print(f"My name is {name}")
print(f"I am {age} years old")
print(f"My score is {score}")
print(f"{name} scored {score}%")
```

```python
# f-strings with expressions
x = 10
y = 20

print(f"Sum: {x + y}")
print(f"Double of x: {x * 2}")
print(f"Uppercase name: {name.upper()}")
```

### Method 2: .format()

```python
name = "Bob"
age = 30

print("My name is {}. I am {} years old".format(name, age))
print("I am {} and I like {}".format(age, "coding"))
```

### Method 3: % operator (Old style)

```python
name = "Charlie"
age = 35

print("My name is %s. I am %d years old" % (name, age))
```

---

## Formatting Numbers

```python
pi = 3.14159

# 2 decimal places
print(f"{pi:.2f}")

# Currency
price = 19.99
print(f"Price: ${price:.2f}")

# Percentage
score = 0.85
print(f"Score: {score:.1%}")

# With commas
num = 1000000
print(f"Number: {num:,}")
```
