# Chapter 4: Lists & Tuples

## What is a List?

A **list** is a collection of items in a specific order. Think of it as a shopping list!

```python
# Creating a list
fruits = ["apple", "banana", "orange", "grape"]
print(fruits)

# A list of numbers
numbers = [1, 2, 3, 4, 5]
print(numbers)

# Mixed types
mixed = ["Hello", 42, 3.14, True]
print(mixed)
```

---

## Accessing List Items

Just like strings, lists have indices (starting from 0)

```python
fruits = ["apple", "banana", "orange", "grape"]

print(fruits[0])    # First item
print(fruits[1])    # Second item
print(fruits[-1])   # Last item
print(fruits[-2])   # Second from last
```

---

## Slicing Lists

```python
numbers = [10, 20, 30, 40, 50, 60]

print(numbers[1:4])   # Items at indices 1, 2, 3
print(numbers[:3])    # First 3 items
print(numbers[3:])    # From index 3 to end
print(numbers[::2])   # Every 2nd item
print(numbers[::-1])  # Reversed!
```

---

## Modifying Lists

Lists are **mutable** - you can change them!

```python
# Change an item
colors = ["red", "blue", "green"]
print("Before:", colors)

colors[1] = "yellow"
print("After:", colors)
```

---

## Adding & Removing Items

```python
tasks = ["Study", "Exercise"]
print("Start:", tasks)

# Add to the end
tasks.append("Code")
print("After append:", tasks)

# Add at specific position
tasks.insert(1, "Eat")
print("After insert:", tasks)
```

```python
tasks = ["Study", "Exercise", "Code", "Sleep"]

# Remove a specific item
tasks.remove("Sleep")
print("After remove:", tasks)

# Remove by index
removed = tasks.pop(0)
print("Removed:", removed)
print("List now:", tasks)
```

---

## List Length

```python
items = ["a", "b", "c", "d", "e"]
print("Number of items:", len(items))

empty_list = []
print("Empty list has:", len(empty_list), "items")
```

---

## What is a Tuple?

A **tuple** is like a list, but **immutable** (cannot be changed)

```python
# Tuples use parentheses
colors = ("red", "green", "blue")
print(colors)

# You can access items like lists
print(colors[0])
print(colors[-1])
print(colors[1:3])
```

```python
# But you cannot modify a tuple
colors = ("red", "green", "blue")

try:
    colors[0] = "yellow"  # This will cause an error
except TypeError as e:
    print("Error:", e)
```

---

## List vs Tuple - When to Use?

| Feature | List | Tuple |
|---------|------|-------|
| Can change? |  Yes |  No |
| Speed | Slower | Faster |
| Syntax | `[1, 2, 3]` | `(1, 2, 3)` |
| Use for | Dynamic data | Fixed data |

## Key Takeaways

 Lists store multiple items in order
 Access items by index: `list[0]`
 Modify lists: `append()`, `remove()`, `insert()`, `pop()`
 Tuples are immutable lists
 Use lists when data changes, tuples when data is fixed
