## Slicing applies to: Strings ✅ · Lists ✅ · Tuples ✅ · Sets ❌ · Dicts ❌

The **same three slicing cases** work identically across strings, lists, and tuples.
Sets and dicts have no positional index, so slicing is not supported for them.

| Case | Expression | start default | end default | step default | Direction |
|------|-----------|--------------|------------|------------|----------|
| End only | `s[:2]` | 0 | 2 | 1 | left → right |
| +ve step, none missing | `s[::2]` | 0 | len(s) | 2 | left → right |
| -ve step, none missing | `s[::-2]` | -1 | -(len+1) | -2 | right → left |

## 1. Lists

A **list** is an ordered, mutable collection. Elements can be of any type.

````python
# Creating lists
cubes  = [1, 8, 27, 64, 125]
mixed  = [1, 'hello', 3.14, True]
nested = [[1, 2], [3, 4]]

print(cubes)
print(mixed)
print(nested)
````

### Indexing

```
 +----+----+----+----+----+
 |  1 |  8 | 27 | 64 |125 |
 +----+----+----+----+----+
    0    1    2    3    4     ← positive index
   -5   -4   -3   -2   -1    ← negative index
```

````python
cubes = [1, 8, 27, 64, 125]

print(cubes[0])    # 1   — first element
print(cubes[4])    # 125 — last element (positive)
print(cubes[-1])   # 125 — last element (negative)
print(cubes[-5])   # 1   — first element (negative)
````

### Slicing — exact same rules as strings

```
list[ start : end : step ]
```

| Part | Default (+ve step) | Default (-ve step) |
|------|-------------------|-------------------|
| `start` | `0` (leftmost) | `-1` (rightmost) |
| `end` | `len(list)` (past last) | `-(len+1)` (past first) |
| `step` | `1` | — |

> 🔑 **+ve step → left to right. -ve step → right to left.**

#### Case 1 — Only `end` given (start and step missing)
- `start` → defaults to **0**
- `step` → defaults to **1** (left → right)

````python
cubes = [1, 8, 27, 64, 125]

print(cubes[:2])    # [1, 8]       — start=0, step=1, first 2 elements
print(cubes[:3])    # [1, 8, 27]   — start=0, step=1, first 3 elements
print(cubes[:0])    # []           — end=0 excludes everything
````

#### Case 2 — Positive step, start and end both missing
**+ve step → reads LEFT → RIGHT**
- `start` → defaults to **0**
- `end` → defaults to **len(list)**

````python
cubes = [1, 8, 27, 64, 125]

# step=1 — every element, left to right → entire list
print(cubes[::1])   # [1, 8, 27, 64, 125]

# step=2 — alternate elements starting from LEFT (index 0, 2, 4 ...)
print(cubes[::2])   # [1, 27, 125]   — 1(0), 27(2), 125(4)

# step=3 — every 3rd element from left (index 0, 3 ...)
print(cubes[::3])   # [1, 64]        — 1(0), 64(3)
````

#### Case 3 — Negative step, start and end both missing
**-ve step → reads RIGHT → LEFT**
- `start` → defaults to **-1** (rightmost element)
- `end` → defaults to **-(len+1)** (goes past the leftmost, so index 0 is included)

````python
cubes = [1, 8, 27, 64, 125]
#
#  +----+----+----+----+----+
#  |  1 |  8 | 27 | 64 |125 |
#  +----+----+----+----+----+
#   -5   -4   -3   -2   -1    ← start from -1 (rightmost) when step is -ve

# step=-1 — every element, right to left → reverses the list
print(cubes[::-1])   # [125, 64, 27, 8, 1]

# step=-2 — every 2nd element from RIGHT (index -1, -3, -5 ...)
print(cubes[::-2])   # [125, 27, 1]   — 125(-1), 27(-3), 1(-5)

# step=-3 — every 3rd element from right
print(cubes[::-3])   # [125, 8]       — 125(-1), 8(-4)

# Equivalent explicit form for cubes[::-2]
print(cubes[-1:-6:-2])  # [125, 27, 1]  — same result
````

````python
# All cases at a glance
cubes = [1, 8, 27, 64, 125]
print(f'{"Expression":<18} {"Result":<20} {"Explanation"}')
print('-' * 65)
cases = [
    ("cubes[:2]",    cubes[:2],    "end only: start=0, step=1"),
    ("cubes[2:]",    cubes[2:],    "start only: end=len, step=1"),
    ("cubes[:]",     cubes[:],     "+ve step, none: full list"),
    ("cubes[::1]",   cubes[::1],   "+ve step 1 from left"),
    ("cubes[::2]",   cubes[::2],   "+ve step 2: alt from left"),
    ("cubes[::3]",   cubes[::3],   "+ve step 3 from left"),
    ("cubes[::-1]",  cubes[::-1],  "-ve step: reversed"),
    ("cubes[::-2]",  cubes[::-2],  "-ve step 2: alt from right"),
    ("cubes[::-3]",  cubes[::-3],  "-ve step 3 from right"),
]
for expr, result, note in cases:
    print(f'{expr:<18} {str(result):<20} {note}')
````

````python
# Lists ARE mutable — you can change elements
cubes[3] = 30
print(cubes)
````

### Adding & Removing Elements

````python
fruits = ['apple', 'banana', 'cherry']

fruits.append('mango')       # Add to end
fruits.insert(1, 'orange')   # Insert at index 1
print(fruits)

fruits.remove('banana')      # Remove by value
popped = fruits.pop()        # Remove and return last element
print(fruits)
print('Popped:', popped)
````

````python
# Slice assignment — powerful list modification
alphabets = ['a', 'b', 'c', 'd', 'e']

alphabets[1:3] = ['B', 'C']  # Replace elements
print(alphabets)

alphabets[1:3] = []          # Delete elements
print(alphabets)

alphabets[:] = []            # Clear the list
print(alphabets)
````

### Useful List Methods

````python
nums = [3, 1, 4, 1, 5, 9, 2, 6]

print(len(nums))      # Length
print(sum(nums))      # Sum
print(min(nums))      # Minimum
print(max(nums))      # Maximum
print(nums.count(1))  # Count occurrences of 1
print(nums.index(5))  # Index of value 5

nums.sort()
print(nums)           # Sort in place

nums.reverse()
print(nums)           # Reverse in place
````

## 2. Tuples

A **tuple** is an ordered, **immutable** collection — like a read-only list.

````python
# Creating tuples
t1 = (1, 2, 3)
t2 = ('apple', 'banana', 'cherry')
t3 = (42,)              # Single-element tuple needs trailing comma!
t4 = ('abc', 34, True)  # Mixed types

print(t1)
print(t3)
print(type(t3))  # <class 'tuple'>
print(type((42)))  # <class 'int'> — no comma = just parentheses
````

````python
# Indexing and slicing — identical rules to lists and strings
fruits = ('apple', 'banana', 'cherry', 'orange', 'kiwi')
print(fruits[0])       # 'apple'
print(fruits[-1])      # 'kiwi'
print(fruits[1:3])     # ('banana', 'cherry')
print(fruits[-3:-1])   # ('cherry', 'orange')

# Case 1 — only end given (start=0, step=1)
print(fruits[:2])      # ('apple', 'banana')

# Case 2 — +ve step, start and end missing → left to right
print(fruits[::1])     # ('apple', 'banana', 'cherry', 'orange', 'kiwi')
print(fruits[::2])     # ('apple', 'cherry', 'kiwi')   — alt from left

# Case 3 — -ve step, start and end missing → right to left
print(fruits[::-1])    # ('kiwi', 'orange', 'cherry', 'banana', 'apple') — reversed
print(fruits[::-2])    # ('kiwi', 'cherry', 'apple')   — alt from right

# Membership test
if 'apple' in fruits:
    print('apple is in the tuple')
````

````python
# Tuples are IMMUTABLE — cannot change elements
try:
    fruits[0] = 'mango'
except TypeError as e:
    print(f'Error: {e}')
````

````python
# Tuple unpacking
point = (10, 20, 30)
x, y, z = point
print(f'x={x}, y={y}, z={z}')

# Swap two values using tuple unpacking
a, b = 5, 10
a, b = b, a
print(f'a={a}, b={b}')
````

## ✅ Quick Summary

| Feature | List | Tuple |
|---------|------|-------|
| Syntax | `[1, 2, 3]` | `(1, 2, 3)` |
| Ordered | ✅ | ✅ |
| Mutable | ✅ | ❌ |
| Duplicates | ✅ | ✅ |
| Slicing | ✅ | ✅ |

**Use a list** when the collection needs to change over time.  
**Use a tuple** for fixed data (coordinates, RGB values, DB rows) — it's faster and hashable.

### Slicing cheat sheet (same for strings, lists, tuples)

```
           step POSITIVE (+)              step NEGATIVE (-)
           ────────────────────           ────────────────────
direction  left  →  right                right  →  left
start      0  (leftmost)                 -1 (rightmost)
end        len(s)  (past last)           -(len+1) (past first)
```
