# Chapter 6: Operators

Every expression in Python — whether it calculates a value, compares two things, or decides how conditions relate — relies on operators. They are the symbols and keywords that tell Python what to do with the data you give it.

## Arithmetic Operators

These are just normal math that you already know.

```python
print("Addition:", 5 + 3)
print("Subtraction:", 10 - 4)
print("Multiplication:", 6 * 7)
print("Division:", 20 / 4)
print("Integer Division:", 17 // 5)  # Divides and rounds down
print("Modulo (remainder):", 17 % 5)
print("Exponent (power):", 2 ** 3)
```

---

## Comparison Operators

A comparison operator examines the relationship between two values and always produces a boolean result — either `True` or `False`. These are the building blocks of every decision your code makes.

```python
print("5 == 5:", 5 == 5)      # Equal
print("5 != 3:", 5 != 3)      # Not equal
print("10 > 5:", 10 > 5)      # Greater than
print("3 < 8:", 3 < 8)        # Less than
print("5 >= 5:", 5 >= 5)      # Greater than or equal
print("7 <= 5:", 7 <= 5)      # Less than or equal
```

---

## Logical Operators

Logical operators join multiple boolean expressions into a single result. Use them when a decision depends on more than one condition at a time.

```python
# AND - both conditions must be true
print("5 > 3 and 10 > 5:", 5 > 3 and 10 > 5)      # True
print("5 > 3 and 10 > 20:", 5 > 3 and 10 > 20)    # False

# OR - at least one must be true
print("5 > 10 or 10 > 5:", 5 > 10 or 10 > 5)      # True

# NOT - reverses the result
print("not True:", not True)                        # False
```

---

## Assignment Operators

Shortcuts for modifying variables.

```python
x = 10
print("Start:", x)

x += 5      # Add to x
print("After += 5:", x)

x -= 3      # Subtract from x
print("After -= 3:", x)

x *= 2      # Multiply x
print("After *= 2:", x)

x /= 2      # Divide x
print("After /= 2:", x)
```

---

## Order Matters

Python does not evaluate an expression strictly left to right. It applies a defined hierarchy: exponentiation binds tightest, followed by multiplication and division, with addition and subtraction resolved last. Parentheses override everything — use them freely to make intent explicit.

```python
# Without parentheses - multiply first
print("2 + 3 * 4:", 2 + 3 * 4)      # 14

# With parentheses - add first
print("(2 + 3) * 4:", (2 + 3) * 4)  # 20

# When in doubt, use parentheses
print("2 ** 3 * 4:", 2 ** 3 * 4)    # 32
```

## Remember

- Arithmetic operators: +, -, *, /, //, %, **
- Comparison operators: ==, !=, >, <, >=, <=
- Logical operators: and, or, not
- Assignment shortcuts: +=, -=, *=, /=
- Order: exponents → multiplication/division → addition/subtraction
