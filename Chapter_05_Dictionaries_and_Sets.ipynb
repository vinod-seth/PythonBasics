## 1. Dictionaries

A **dictionary** maps each **key** to a corresponding **value**, giving you a way to label and retrieve data by name rather than by position. Keys must be unique within a dictionary.

Unlike a list — where you access items by their numeric index — a dictionary lets you use meaningful identifiers as keys, making it far easier to work with structured data like records, configurations, or lookup tables.

```python
# Creating a dictionary
car = {
    'brand': 'Maruti',
    'model': 'Swift',
    'year' : 2024
}
print(car)
```

### Accessing Values

```python
print(car['brand'])          # Direct access — raises KeyError if missing
print(car.get('model'))      # .get() — returns None if key missing (safe)
print(car.get('color', 'Unknown'))  # Default value if key missing
```

### Updating & Adding Items

```python
car['brand'] = 'Hyundai'          # Update existing key
car.update({'year': 2025})         # Update using .update()
car['autodrive'] = 'enabled'       # Add a new key
print(car)
```

### Removing Items

```python
removed = car.pop('model')         # Remove by key, returns value
print('Removed:', removed)

del car['year']                    # Delete with del keyword
print(car)
```

### Iterating Over a Dictionary

```python
person = {'name': 'Alice', 'age': 30, 'country': 'India'}

print(person.keys())    # All keys
print(person.values())  # All values
print(person.items())   # Key-value pairs as tuples

# Iterate with items()
for key, value in person.items():
    print(f'{key} -> {value}')
```

### Copying a Dictionary

> Writing `b = a` does **not** produce a second dictionary — it creates a second name for the **same** underlying object. Any change made through `b` will be visible through `a`, because there is only one dictionary in memory.

```python
original = {'x': 1, 'y': 2}

# ❌ This is a REFERENCE, not a copy
ref = original
ref['x'] = 99
print('Original:', original)   # original is also changed!

# ✅ Proper copies
original = {'x': 1, 'y': 2}
copy1 = original.copy()
copy2 = dict(original)

copy1['x'] = 99
print('original:', original)   # unchanged
print('copy1:', copy1)
```

### Nested Dictionaries

```python
people = {
    'Alice': {'age': 30, 'country': 'India'},
    'Bob':   {'age': 25, 'country': 'UK'},
}

for name, info in people.items():
    print(name)
    for key, val in info.items():
        print(f'  {key}: {val}')
```

## 2. Sets

A **set** holds only distinct values and imposes no fixed order on them. If you add a duplicate, Python silently discards it — the set simply keeps one copy.

> **Sets and dicts do not support indexing or slicing** — neither structure tracks element positions, so `s[0]` on a set raises a `TypeError`. To work through the contents in order, either iterate with a `for` loop or convert to a list first.

```python
# Creating sets
s1 = {'apple', 'banana', 'cherry'}
s2 = set(['apple', 'cherry', 'banana', 'apple'])  # Duplicates removed

print(s1)
print(s2)  # Only unique values
```

```python
# Set operations
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b)   # Union — all elements
print(a & b)   # Intersection — common elements
print(a - b)   # Difference — in a but not in b
print(a ^ b)   # Symmetric difference — in one but not both
```

```python
# Add and remove
s1.add('mango')
s1.discard('apple')  # No error if element doesn't exist
print(s1)
```

##  Quick Summary

| Feature | Dictionary | Set |
|---------|:----------:|:---:|
| Syntax | `{'k': v}` | `{1, 2, 3}` |
| Ordered | ✓ (3.7+) | ✗ |
| Mutable | ✓ | ✓ |
| Duplicate values | Values: ✓, Keys: ✗ | ✗ |
| Slicing | ✗ | ✗ |
| Access | `d['key']` or `d.get('key')` | membership: `x in s` |

**Use a dict** when you need to look up data by a meaningful name (key).  
**Use a set** when you need uniqueness or fast membership tests and set algebra (union, intersection, difference).
