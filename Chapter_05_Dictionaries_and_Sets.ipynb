# Chapter 5: Dictionaries and Sets

## What is a Dictionary?

A dictionary is like a real dictionary - you look up a word (key) to get its definition (value). It's perfect for storing related information together.

```python
# Creating a dictionary
student = {
    "name": "Alice",
    "age": 20,
    "grade": "A"
}

print(student)
```

---

## Getting Information from a Dictionary

```python
student = {
    "name": "Alice",
    "age": 20,
    "grade": "A"
}

# Access by key
print(student["name"])
print(student["age"])
print(student["grade"])

# get() is safer if key might not exist
print(student.get("name"))
print(student.get("email", "Not found"))
```

---

## Changing Dictionaries

```python
student = {"name": "Bob", "age": 19}
print("Start:", student)

# Update a value
student["age"] = 20
print("After update:", student)

# Add a new key-value pair
student["grade"] = "B"
print("After adding:", student)

# Remove a pair
del student["grade"]
print("After removal:", student)
```

---

## Dictionary Methods

```python
person = {
    "name": "Charlie",
    "age": 25,
    "city": "NYC"
}

# Get all keys
print("Keys:", person.keys())

# Get all values
print("Values:", person.values())

# Get key-value pairs
print("Items:", person.items())
```

---

## What is a Set?

A set is a collection where each item appears only once. Duplicates are automatically removed. Think of it like a group of unique items.

```python
# Creating a set
colors = {"red", "blue", "green"}
print(colors)

# If you add duplicates, they're ignored
numbers = {1, 2, 2, 3, 3, 3}
print(numbers)  # Notice: only 1, 2, 3
```

```python
# Adding and removing from sets
fruits = {"apple", "banana"}
print("Start:", fruits)

fruits.add("orange")
print("After add:", fruits)

fruits.remove("banana")
print("After remove:", fruits)
```

---

## Set Operations

```python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

# Union - all unique items from both
print("Union:", set1 | set2)

# Intersection - items in both
print("Intersection:", set1 & set2)

# Difference - items in set1 but not set2
print("Difference:", set1 - set2)
```

---

## Checking if Something Exists

```python
# Checking in a dictionary
person = {"name": "David", "age": 30}
print("name" in person)   # True
print("email" in person)  # False

# Checking in a set
colors = {"red", "blue", "green"}
print("red" in colors)     # True
print("yellow" in colors)  # False
```

---

## When to Use What?

| Type | Use For | Example |
|------|---------|----------|
| List | Ordered items that can change | `[1, 2, 3]` |
| Dictionary | Related information (key-value) | `{"name": "Alice"}` |
| Set | Unique items | `{1, 2, 3}` |
