## 1. The `while` Loop

A `while` loop keeps executing its body as long as a given condition evaluates to `True`. Once the condition becomes `False`, execution moves past the loop.

````python
# Basic while loop
i = 1
while i <= 5:
    print(i)
    i += 1
````

````python
# break — exit the loop early
i = 1
while i <= 10:
    if i == 4:
        break      # Stop when i reaches 4
    print(i)
    i += 1

print('Loop ended')
````

````python
# else — runs when condition becomes False (not when break is hit)
i = 1
while i < 4:
    print(i)
    i += 1
else:
    print('Loop finished normally, i =', i)
````

## 2. The `for` Loop

A `for` loop steps through the elements of any sequence or iterable — such as a list, string, dictionary, or `range` — processing each item in turn.

````python
# Loop over a list
cars = ['Maruti', 'Audi', 'Mercedes', 'Ferrari']
for car in cars:
    print(car)
````

````python
# Loop over a string
for ch in 'Python':
    print(ch, end=' ')
print()  # newline
````

````python
# range() — generate a sequence of numbers
for i in range(5):          # 0, 1, 2, 3, 4
    print(i, end=' ')
print()

for i in range(2, 10, 2):   # 2, 4, 6, 8
    print(i, end=' ')
print()
````

````python
# enumerate() — get index and value together
fruits = ['apple', 'banana', 'cherry']
for i, fruit in enumerate(fruits):
    print(f'{i}: {fruit}')
````

````python
# continue — skip current iteration and move to next
cars = ['Maruti', 'Audi', 'Mercedes', 'Ferrari']
for car in cars:
    if car == 'Audi':
        continue    # Skip Audi
    print(car)
````

````python
# break — stop the loop early
for car in cars:
    print(car)
    if car == 'Mercedes':
        break    # Stop after Mercedes
````

## 3. Functions

A **function** is a named unit of logic: you write the steps once, give that group a name, and then invoke it wherever you need those steps — with `def` to define it and the name followed by parentheses to call it.

````python
# Basic function
def greet(name):
    """Say hello to someone."""
    return f'Hello, {name}!'

print(greet('Alice'))
print(greet('Bob'))
````

````python
# Default parameter values
def greet(name, greeting='Hello'):
    return f'{greeting}, {name}!'

print(greet('Alice'))              # Uses default greeting
print(greet('Bob', 'Good morning'))  # Override default
````

````python
# Multiple return values
def min_max(numbers):
    return min(numbers), max(numbers)

lo, hi = min_max([3, 1, 7, 4, 9, 2])
print(f'Min: {lo}, Max: {hi}')
````

## 4. `*args` and `**kwargs`

Sometimes you do not know ahead of time how many arguments a function will receive. `*args` collects any number of extra **positional** values into a **tuple**, while `**kwargs` gathers any extra **keyword** arguments into a **dict**.

````python
# Fixed args — can only pass exactly 3 numbers
def add_three(x, y, z):
    return x + y + z

print(add_three(1, 2, 3))
````

````python
# *args — accept any number of positional arguments
def add(*nums):           # nums is a tuple of all arguments
    return sum(nums)

print(add(3, 5))          # 8
print(add(1, 2, 3, 4, 5)) # 15
print(add(10))            # 10
````

````python
# **kwargs — accept any number of keyword arguments
def introduce(**data):    # data is a dict
    print('Type:', type(data))
    for key, value in data.items():
        print(f'  {key}: {value}')

introduce(Firstname='Sita', Lastname='Sharma', Country='India', Age=28)
````

````python
# Combining: regular, *args, **kwargs
def demo(required, *args, **kwargs):
    print('Required:', required)
    print('args:', args)
    print('kwargs:', kwargs)

demo('hello', 1, 2, 3, name='Alice', city='Delhi')
````

## 5. Lambda Functions

A **lambda** expression defines a function inline, without a `def` statement or a name — useful when you need a short throwaway computation in a single expression.

```python
lambda arguments : expression
```

````python
# Regular function vs lambda
def square(n):
    return n ** 2

square_lambda = lambda n: n ** 2

print(square(5))         # 25
print(square_lambda(5))  # 25 — same result
````

````python
# Lambda with multiple arguments
add = lambda x, y: x + y
print(add(3, 7))   # 10
````

````python
# Lambdas shine with map(), filter(), sorted()
numbers = [1, 2, 3, 4, 5]

# map() — apply function to every element
squares = list(map(lambda x: x ** 2, numbers))
print('Squares:', squares)

# filter() — keep elements where function returns True
evens = list(filter(lambda x: x % 2 == 0, numbers))
print('Evens:', evens)

# sorted() with custom key
words = ['banana', 'apple', 'cherry', 'kiwi']
by_length = sorted(words, key=lambda w: len(w))
print('By length:', by_length)
````

## 6. List Comprehensions

List comprehensions let you build a new list by expressing the transformation and optional filtering rule in a single, readable line — rather than writing an explicit loop with `append`.

```python
[expression for item in iterable if condition]
```

````python
# Traditional approach
squares = []
for x in range(10):
    squares.append(x ** 2)
print(squares)
````

````python
# List comprehension — same result, one line!
squares = [x ** 2 for x in range(10)]
print(squares)
````

````python
# With a condition (filter)
even_squares = [x ** 2 for x in range(10) if x % 2 == 0]
print(even_squares)
````

````python
# String manipulation
fruits = ['apple', 'banana', 'cherry', 'avocado', 'kiwi']

# Uppercase of fruits starting with 'a'
a_fruits = [f.upper() for f in fruits if f.startswith('a')]
print(a_fruits)
````

````python
# Dict comprehension
word_lengths = {word: len(word) for word in fruits}
print(word_lengths)

# Set comprehension
unique_lengths = {len(word) for word in fruits}
print(unique_lengths)
````
