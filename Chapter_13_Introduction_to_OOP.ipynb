## What is OOP?

**OOP** is a way to organize code using **objects** and **classes**.

Think of a class as a **recipe** and objects as the **individual meals** produced from it — the recipe defines the ingredients and steps, while each meal is a distinct result you can serve, modify, and consume independently.

---

## Creating Your First Class

```python
# Define a class
class Dog:
    # Constructor - runs when creating object
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    # Method - function inside class
    def bark(self):
        print(f"{self.name} says: Woof! Woof!")
    
    def get_info(self):
        return f"{self.name} is {self.age} years old"

# Create objects (instances)
dog1 = Dog("Buddy", 3)
dog2 = Dog("Max", 5)

# Use objects
dog1.bark()
dog2.bark()
print(dog1.get_info())
print(dog2.get_info())
```

---

## Understanding self

`self` refers to the specific object calling the method

```python
class Car:
    def __init__(self, brand, color):
        self.brand = brand  # This object's brand
        self.color = color  # This object's color
    
    def describe(self):
        print(f"I am a {self.color} {self.brand}")

car1 = Car("Toyota", "blue")
car2 = Car("Honda", "red")

car1.describe()  # self = car1
car2.describe()  # self = car2
```

## Attributes and Methods

```python
class Student:
    def __init__(self, name, grade):
        self.name = name      # Attribute
        self.grade = grade    # Attribute
    
    # Method - can use and modify attributes
    def improve_grade(self):
        if self.grade < 100:
            self.grade += 5
        print(f"{self.name}'s new grade: {self.grade}")
    
    def get_status(self):
        if self.grade >= 90:
            status = "Excellent"
        elif self.grade >= 80:
            status = "Good"
        else:
            status = "Average"
        return f"{self.name}: {status}"

student = Student("Alice", 85)
print(student.get_status())
student.improve_grade()
print(student.get_status())
```

## Creating Multiple Objects

```python
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages
    
    def info(self):
        return f"{self.title} by {self.author} ({self.pages} pages)"

# Create multiple books
books = [
    Book("Python 101", "Madui Smth", 300),
    Book("Web Dev Guide", "Jane Doe", 450),
    Book("Data Science", "Wil Boby", 520)
]

# Print all books
for book in books:
    print(book.info())
```

## Bank Account Example

```python
class BankAccount:
    def __init__(self, holder, balance=0):
        self.holder = holder
        self.balance = balance
    
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount}. New balance: ${self.balance}")
        else:
            print("Deposit amount must be positive")
    
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds!")
        elif amount > 0:
            self.balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.balance}")
        else:
            print("Withdrawal amount must be positive")
    
    def check_balance(self):
        print(f"{self.holder}'s balance: ${self.balance}")

# Use the class
account = BankAccount("Madui", 1000)
account.check_balance()
account.deposit(500)
account.withdraw(200)
account.check_balance()
```

## Check Object Type

```python
class Person:
    def __init__(self, name):
        self.name = name

person = Person("Boby")

# Check type
print(type(person))
print(isinstance(person, Person))  # True
print(isinstance(person, str))     # False
```
