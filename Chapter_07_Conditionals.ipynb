# Chapter 7: Conditional Statements

## What are Conditionals?

**Conditionals** let your code make decisions. "If this happens, do that!"

---

## The if Statement

```python
age = 18

if age >= 18:
    print("You are an adult!")
    print("Welcome!")
```

**Important:** Notice the colon `:` and indentation (4 spaces)

```python
score = 45

if score >= 50:
    print("You passed!")

print("Test completed")  # This prints no matter what
```

---

## The if-else Statement

Do something if True, otherwise do something else

```python
age = 15

if age >= 18:
    print("You are an adult")
else:
    print("You are a teenager")
```

```python
temperature = 20

if temperature > 25:
    print("It's hot! ☀️")
else:
    print("It's cool 🌤️")
```

---

## The if-elif-else Statement

Make multiple decisions

```python
score = 75

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score}, Grade: {grade}")
```

---

## Using Logical Operators in Conditions

```python
age = 25
has_license = True

# AND - both must be True
if age >= 18 and has_license:
    print("You can drive!")
```

```python
day = "Saturday"

# OR - at least one must be True
if day == "Saturday" or day == "Sunday":
    print("It's the weekend! ")
else:
    print("Time for work ")
```

```python
is_raining = False

# NOT - reverses the condition
if not is_raining:
    print("Let's go outside!")
```

---

## Nested Conditions

Conditions inside conditions

```python
age = 22
has_money = True

if age >= 18:
    print("You are an adult")
    
    if has_money:
        print("You can buy a ticket!")
    else:
        print("You need money for a ticket")
else:
    print("You are too young")
```

---

## Checking in Collections

```python
# Check if item is in list
fruits = ["apple", "banana", "orange"]

if "apple" in fruits:
    print("Apple is available")

if "grape" in fruits:
    print("Grape is available")
else:
    print("Grape is NOT available")
```

```python
# Check in dictionary
person = {"name": "Alice", "age": 25, "city": "NYC"}

if "name" in person:
    print("Name is", person["name"])

if "email" in person:
    print(person["email"])
else:
    print("Email not available")
```
