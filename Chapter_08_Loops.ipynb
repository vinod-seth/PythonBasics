# Chapter 8: Loops

## What are Loops?

**Loops** let you repeat code without writing it over and over

---

## The for Loop

Loop **for** a specific number of times or through items

```python
# Loop through a list
fruits = ["apple", "banana", "orange"]

for fruit in fruits:
    print(fruit)
```

```python
# Use range() to loop a number of times
for i in range(5):
    print(f"Number {i}")
```

**Note:** `range(5)` generates: 0, 1, 2, 3, 4 (stops before 5)

```python
# range with start and end
for i in range(3, 8):  # 3, 4, 5, 6, 7
    print(i)
```

```python
# range with step
for i in range(0, 10, 2):  # Every 2
    print(i)
```

---

## Looping Through Different Collections

```python
# Through a string
word = "Python"

for letter in word:
    print(letter)
```

```python
# Through a tuple
colors = ("red", "green", "blue")

for color in colors:
    print(f"Color: {color}")
```

```python
# Through a dictionary (keys)
person = {"name": "Alice", "age": 25, "city": "NYC"}

for key in person:
    print(f"{key}: {person[key]}")
```

```python
# Through a set
numbers = {1, 2, 3, 4, 5}

for num in numbers:
    print(num * 2)
```

---

## The while Loop

**While** a condition is True, keep looping

```python
count = 0

while count < 5:
    print(f"Count: {count}")
    count += 1  # Important! Change count or loop forever!
```

```python
# Guessing game example
secret = 5
guess = 0

while guess != secret:
    print("Wrong! Try again.")
    guess += 1

print("You got it! The number was", secret)
```

---

## Loop Control: break

**break** exits the loop immediately

```python
for i in range(10):
    if i == 5:
        break  # Stop the loop
    print(i)
```

```python
# Find an item in a list
numbers = [10, 20, 30, 40, 50]
search = 30

for num in numbers:
    if num == search:
        print(f"Found {search}!")
        break
    print(f"Checking {num}...")
```

---

## Loop Control: continue

**continue** skips to the next iteration

```python
for i in range(5):
    if i == 2:
        continue  # Skip this iteration
    print(i)
```

```python
# Skip even numbers
for i in range(1, 11):
    if i % 2 == 0:  # If even
        continue
    print(f"Odd: {i}")
```

---

## Nested Loops

Loops inside loops

```python
# Times table
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i}  {j} = {i*j}")
    print()  # Empty line between rows
```

```python
# Simple pattern
for i in range(1, 5):
    for j in range(i):
        print("*", end="")
    print()  # New line
```

---

## Practical Examples

```python
# Sum all numbers in a list
numbers = [10, 20, 30, 40, 50]
total = 0

for num in numbers:
    total += num

print(f"Sum: {total}")
```

```python
# Find maximum value
scores = [75, 92, 88, 95, 78]
max_score = scores[0]

for score in scores:
    if score > max_score:
        max_score = score

print(f"Highest score: {max_score}")
```

 `for`: Loop through items or a range

 `break`: Exit loop
 `continue`: Skip to next iteration
