## 1. What is a Module?

A **module** is simply a `.py` file containing Python code (functions, classes, variables).  
Modules let you **organize** and **reuse** code across different programs.

Python has three types of modules:
1. **Built-in** — included in Python itself (e.g., `math`, `os`, `sys`)
2. **Third-party** — installed via pip (e.g., `pandas`, `numpy`)
3. **Your own** — `.py` files you write

---
## 2. Importing Modules

````python
# Import entire module
import math

print(math.pi)              # 3.14159...
print(math.sqrt(16))        # 4.0
print(math.factorial(5))    # 120
````

````python
# Import specific names from a module
from math import pi, sqrt, ceil

print(pi)
print(sqrt(25))
print(ceil(4.3))
````

````python
# Import with an alias
import math as m

print(m.pi)
print(m.floor(4.9))
````

````python
# Explore common built-in modules
import os
import sys
import random
import datetime

print(os.getcwd())                          # Current working directory
print(random.randint(1, 100))               # Random integer
print(datetime.date.today())                # Today's date
````

## 3. The `dir()` Function

`dir()` returns a sorted list of all names defined in a module (or object).

````python
import math

# See all names in the math module
print(dir(math))
````

````python
# Filter out private/dunder names
public_names = [name for name in dir(math) if not name.startswith('_')]
print(public_names)
````

````python
# dir() on any object shows its methods and attributes
s = 'hello'
string_methods = [m for m in dir(s) if not m.startswith('_')]
print(string_methods)
````

````python
# __name__ — the module's name as a string
import math
print(math.__name__)   # 'math'
print(__name__)        # '__main__' when run directly
````

## 4. Module Search Path (`sys.path`)

When you write `import spam`, Python searches for it in this order:
1. The current directory
2. Directories in the `PYTHONPATH` environment variable
3. Standard library directories
4. Site-packages (installed third-party packages)

All these directories are listed in `sys.path`.

````python
import sys

print('Python version:', sys.version)
print('\nModule search path:')
for path in sys.path:
    print(' ', path)
````

````python
# You can add to sys.path at runtime
sys.path.append('/my/custom/directory')
print(sys.path[-1])  # Verify it was added
````

## 5. Creating Your Own Module

Any `.py` file is a module. Here we simulate creating and using one.

````python
# Simulate writing a module file: mymath.py
module_code = '''
# mymath.py
PI = 3.14159

def area_circle(r):
    """Return area of a circle."""
    return PI * r * r

def area_rect(w, h):
    """Return area of a rectangle."""
    return w * h
'''

with open('mymath.py', 'w') as f:
    f.write(module_code)

print('Module file written!')
````

````python
# Now import and use it
import mymath

print(mymath.PI)
print(mymath.area_circle(5))
print(mymath.area_rect(4, 6))
````

````python
# Import specific names
from mymath import area_circle, PI

print(area_circle(3))
print(PI)
````

## 6. Packages and `__init__.py`

A **package** is a directory of modules. It must contain an `__init__.py` file (can be empty) to be recognized as a Python package.

```
mypackage/
    __init__.py      ← marks this folder as a package
    module_a.py
    module_b.py
```

````python
import os

# Create a package structure
os.makedirs('mypackage', exist_ok=True)

# __init__.py — controls what 'from mypackage import *' exposes
with open('mypackage/__init__.py', 'w') as f:
    f.write("__all__ = ['greetings', 'utils']\n")

with open('mypackage/greetings.py', 'w') as f:
    f.write('def hello(name):\n    return f"Hello, {name}!"\n')

with open('mypackage/utils.py', 'w') as f:
    f.write('def add(a, b):\n    return a + b\n')

print('Package created!')
````

````python
# Import from a package
from mypackage.greetings import hello
from mypackage.utils import add

print(hello('Alice'))
print(add(3, 7))
````

## 8. Useful Built-in Modules

| Module | Purpose |
|--------|----------|
| `math` | Mathematical functions (sqrt, ceil, floor, pi) |
| `os` | OS interaction (paths, directories, environment) |
| `sys` | Interpreter info, sys.path, sys.argv |
| `random` | Random numbers and choices |
| `datetime` | Dates, times, timedeltas |
| `json` | Parse and serialize JSON |
| `re` | Regular expressions |
| `pathlib` | Object-oriented filesystem paths |

````python
# os — working with paths
import os
from pathlib import Path

print(os.getcwd())                  # Current directory
print(os.path.join('folder', 'file.txt'))  # Path join

# pathlib — modern, object-oriented way
p = Path('.') / 'conf' / 'config.yml'
print(str(p))                       # './conf/config.yml'
print(Path.home())                  # Home directory
````

````python
# json — serialize Python objects
import json

data = {'name': 'Alice', 'scores': [95, 87, 92], 'active': True}

json_str = json.dumps(data, indent=2)  # Python → JSON string
print(json_str)

parsed = json.loads(json_str)          # JSON string → Python
print(parsed['name'])
````
