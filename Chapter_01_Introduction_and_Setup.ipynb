## Why Learn Python?

If you're starting to code, Python is a strong first language:

- Its syntax is clean and close to plain English, reducing the mental overhead of learning
- Employers across tech, finance, and science actively seek Python skills
- A single language lets you build web services, process data, develop AI models, and automate repetitive work
- An active global community means well-maintained libraries and answers to nearly every question you'll have

## Your First Python Program

Let's write the classic "Hello, World" program - the first program every programmer writes:

```python
print("Hello, World!")
```

Congratulations! You just wrote Python code. Here's what happened:

- `print()` is a function that displays text on the screen
- `"Hello, World!"` is the text we want to show
- When you run this code, Python executes it and displays the output

## Let's Print Some More Things

```python
# This is a comment. Python ignores everything after the #
print("My name is Madui")
print("I am learning Python")
print("This language is awesome")
```

Notice two things:

- Each `print()` statement creates a new line
- Lines starting with # are comments - notes for humans that Python completely ignores

## Python Can Do Math

```python
print(5 + 3)
print(10 - 2)
print(4 * 5)
print(20 / 4)
```

Python can perform basic arithmetic. It reads the code from top to bottom and executes each line.
