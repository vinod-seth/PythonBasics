# Chapter 3: Strings - Text Mastery

## Creating Strings

Strings are text data. You create them using **single or double quotes**.

```python
# Both work the same way
name1 = 'Python'
name2 = "Python"

print(name1)
print(name2)
print(name1 == name2)  # Are they equal?
```

---

## Quotes Inside Strings

What if you want quotes inside your string? Use the opposite type!

```python
# Single quotes inside double quotes
sentence1 = "He said 'Hello!'"
print(sentence1)

# Double quotes inside single quotes
sentence2 = 'She said "Goodbye!"'
print(sentence2)
```

---

## Combining Strings (Concatenation)

Join strings together using `+`

```python
first_name = "John"
last_name = "Doe"

# Combine them
full_name = first_name + " " + last_name
print(full_name)
```

```python
# You can also multiply strings!
laugh = "Ha" * 3
print(laugh)

dashes = "-" * 10
print(dashes)
```

---

## String Length

Find how many characters are in a string using `len()`

```python
word = "Python"
print("The word is:", word)
print("Number of characters:", len(word))

# Length of longer strings
sentence = "I love Python programming"
print("The sentence has", len(sentence), "characters (including spaces)")
```

---

## Indexing - Getting Characters

Each character has a position (index). Python starts counting from **0**!

```python
word = "Python"

# Indexing illustration:
# P y t h o n
# 0 1 2 3 4 5  (forward)
#-6-5-4-3-2-1  (backward)

print("First character:", word[0])  # P
print("Second character:", word[1]) # y
print("Last character:", word[-1]) # n
print("Second from last:", word[-2]) # o
```

---

## Slicing - Getting Substrings

Get a "slice" of a string using `[start:end]`

```python
word = "Python"

# word[0:2] - from index 0 up to (not including) index 2
print(word[0:2])   # "Py"

# word[2:5] - from index 2 up to index 5
print(word[2:5])   # "tho"

# word[:3] - from start up to index 3
print(word[:3])    # "Pyt"

# word[3:] - from index 3 to the end
print(word[3:])    # "hon"
```

```python
# Slicing with negative indices
word = "Python"

print(word[-4:-1])  # "tho"
print(word[:-2])    # "Pyth"
print(word[-3:])    # "hon"
```

---

## Slicing with Step

Skip characters using `[start:end:step]`

```python
word = "Python"

# Every 2nd character
print(word[0:6:2])  # "Pto"

# Reverse a string!
print(word[::-1])   # "nohtyP"

# Every 2nd character in reverse
print(word[::-2])   # "nhy"
```

---

## Checking If Text Exists

Use `in` to check if something is in a string

```python
sentence = "I love learning Python"

print("love" in sentence)      # True
print("Java" in sentence)      # False
print("Python" in sentence)    # True
```
