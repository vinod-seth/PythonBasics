## 1. Creating Strings

Strings can be created using **single quotes**, **double quotes**, or **triple quotes**.
- Single and double quotes are interchangeable for simple strings.
- Triple quotes allow multi-line strings.

````python
s1 = 'Hello'
s2 = "World"
s3 = '''This is
a multi-line
string'''

print(s1)
print(s2)
print(s3)
````

### Embedding Quotes Inside Strings

Use the opposite quote type to embed quotes, or use a backslash `\` to escape.

````python
s4 = "Hello how ' are you"   # double quotes hide single quote
s5 = '"Yes," he said.'        # single quotes hide double quote
s6 = "She said \"hello\""     # escape with backslash

print(s4)
print(s5)
print(s6)
````

## 2. String Indexing

A string is an **ordered collection of characters** — each character occupies a numbered slot, much like keys on a keyboard laid out in a row. Python assigns every slot two labels: a **positive index** counting from the left, and a **negative index** counting from the right.

```
  word:   l  a  p  t  o  p
          |  |  |  |  |  |
pos idx:  0  1  2  3  4  5
neg idx: -6 -5 -4 -3 -2 -1
```

Both labels refer to the same character — you choose whichever is more convenient.

````python
word = 'laptop'

print(word[0])    # First character
print(word[5])    # Last character (positive index)
print(word[-1])   # Last character (negative index)
print(word[-6])   # First character (negative index)
````

````python
# Accessing an index that doesn't exist raises IndexError
try:
    print(word[30])
except IndexError as e:
    print(f"Error: {e}")
````

## 3. String Slicing

Pull out a portion of a string using the slice syntax:

```
string[ start : end : step ]
```

Think of `start` and `end` as mile-markers on a road, and `step` as the gear that sets both your **direction** and how far you jump between stops.

| Part | Default (+ve step) | Default (-ve step) | Meaning |
|------|-------------------|-------------------|---------|
| `start` | `0` (leftmost) | `-1` (rightmost) | First position visited |
| `end` | `len(string)` (past last) | `-len-1` (past first) | Boundary position — never included ✗ |
| `step` | `1` | — | Gap between picks; **+ve → left to right**, **-ve → right to left** |

> **Core principle:** the **sign of `step`** sets the **direction of travel**, and all omitted defaults are chosen to match that direction automatically.

````python
#  +---+---+---+---+---+---+
#  | P | y | t | h | o | n |
#  +---+---+---+---+---+---+
#    0   1   2   3   4   5    ← positive index
#   -6  -5  -4  -3  -2  -1   ← negative index

word = 'Python'
````

### Case 1 — Only `end` given (start and step missing)

When only `end` is supplied, Python fills in the two missing pieces with their positive-step defaults:

- `start` not provided → falls back to **0** (the very first character on the left)
- `step` not provided → falls back to **1** (advance one position at a time, left to right)

```
word[:2]  →  start=0, end=2, step=1  →  collects index 0, 1  →  'Py'
```

````python
word = 'Python'

print(word[:2])   # start=0, step=1  → 'Py'    (first 2 chars)
print(word[:4])   # start=0, step=1  → 'Pyth'  (first 4 chars)
print(word[:0])   # end=0 excludes everything → ''

# Explicit equivalent
print(word[0:2:1])  # same as word[:2]
````

### Case 2 — Positive step, both start and end missing

**+ve step → traversal goes LEFT to RIGHT**

With a positive step and nothing else specified, Python picks the widest possible left-to-right range:

- `start` not provided → defaults to **0** (leftmost position)
- `end` not provided → defaults to **len(string)** (one slot past the final character)

```
word[::1]  →  start=0, end=6, step=1   →  full string, left to right
word[::2]  →  start=0, end=6, step=2   →  every other character from the left
word[::3]  →  start=0, end=6, step=3   →  every 3rd character from the left
```

````python
word = 'Python'

# step=1 → every character, left to right
print(word[::1])   # 'Python'  — entire string (identical to word[:])

# step=2 → alternate characters starting from left (index 0, 2, 4 ...)
print(word[::2])   # 'Pto'     — P(0), t(2), o(4)

# step=3 → every 3rd character from left (index 0, 3 ...)
print(word[::3])   # 'Ph'      — P(0), h(3)

# With explicit start (step still +ve)
print(word[1::2])  # 'yhn'    — y(1), h(3), n(5)
````

### Case 3 — Negative step, both start and end missing

**-ve step → traversal goes RIGHT to LEFT**

With a negative step and nothing else specified, Python picks the widest possible right-to-left range:

- `start` not provided → defaults to **-1** (rightmost character)
- `end` not provided → defaults to **-(len+1)**, a sentinel position just beyond the leftmost character, ensuring index 0 is included ✓

```
word[::-1]  →  start=-1, end=-(6+1), step=-1  →  full string, right to left  →  'nohtyP'
word[::-2]  →  start=-1, end=-(6+1), step=-2  →  every other character from the right  →  'nhy'
```

````python
word = 'Python'

#  +---+---+---+---+---+---+
#  | P | y | t | h | o | n |
#  +---+---+---+---+---+---+
#   -6  -5  -4  -3  -2  -1     ← start from -1 (rightmost)

# step=-1 → every character, right to left → reverses the string
print(word[::-1])   # 'nohtyP'  — n(-1), o(-2), h(-3), t(-4), y(-5), P(-6)

# step=-2 → every 2nd character from right
print(word[::-2])   # 'nhy'     — n(-1), h(-3), y(-5)

# step=-3 → every 3rd character from right
print(word[::-3])   # 'nt'      — n(-1), t(-4)

# Equivalent explicit form for word[::-2]
# start=-1, stop must go past index 0 going leftward, so stop=-7
print(word[-1:-7:-2])  # 'nhy'  — same result
````

### Case 4 — All three specified

Explicit `start`, `end`, and `step` — full control.

````python
word = 'Python'

print(word[0:4:1])   # 'Pyth'  — index 0,1,2,3 (end=4 excluded)
print(word[0:6:2])   # 'Pto'   — index 0,2,4
print(word[2:5:1])   # 'tho'   — index 2,3,4
print(word[5:1:-1])  # 'noht'  — index 5,4,3,2 (right to left, end=1 excluded)
print(word[5:0:-2])  # 'nt'    — index 5,3,1
````

### Common Gotcha — mismatched step direction and index order

Imagine you set your car's gear to **drive forward**, but your destination is **behind** you. You will never reach it — you simply stay put. Slicing works the same way:

- `step` is **+ve** (move left to right), but `start` sits to the **right** of `end` → Python has nowhere to go and returns **`''`** ✗
- `step` is **-ve** (move right to left), but `start` sits to the **left** of `end` → same impossibility, result is **`''`** ✗

The step direction and the order of your start/end positions must be consistent, or you get an empty result.

````python
word = 'Python'

# step=+1 but start(4) > end(1) — mismatch → empty
print(repr(word[4:1:1]))    # ''

# step=-1 but start(1) < end(4) — mismatch → empty
print(repr(word[1:4:-1]))   # ''

# Correct: step=-1 with start(4) > end(1) → works
print(word[4:1:-1])         # 'oht'  — index 4,3,2
````

### All cases at a glance

````python
word = 'Python'
print(f'{"Expression":<18} {"Result":<12} {"Explanation"}')
print('-' * 60)

cases = [
    ("word[:2]",    word[:2],    "end only: start=0, step=1"),
    ("word[2:]",    word[2:],    "start only: end=len, step=1"),
    ("word[:]",     word[:],     "+ve step, none given: full string"),
    ("word[::1]",   word[::1],   "+ve step 1 from left"),
    ("word[::2]",   word[::2],   "+ve step 2: alternate from left"),
    ("word[::3]",   word[::3],   "+ve step 3 from left"),
    ("word[::-1]",  word[::-1],  "-ve step: reverse (start=-1)"),
    ("word[::-2]",  word[::-2],  "-ve step 2: alternate from right"),
    ("word[::-3]",  word[::-3],  "-ve step 3 from right"),
]
for expr, result, note in cases:
    print(f'{expr:<18} {str(result):<12} {note}')
````

### Out-of-range slices — no error!

````python
word = 'Python'

# Out-of-range slices are clipped gracefully (unlike single index access)
print(word[2:100])   # 'thon' — clips at end
print(word[50:])     # ''     — start beyond string → empty
print(word[-100:2])  # 'Py'   — start clipped to 0

# But single index out-of-range → IndexError!
try:
    print(word[30])
except IndexError as e:
    print(f'IndexError: {e}')
````

## 4. String Immutability

Strings in Python **cannot be changed** after creation — they are **immutable**. To 'modify' a string, you must create a new one.

````python
word = 'Python'

# This will raise a TypeError
try:
    word[0] = 'J'
except TypeError as e:
    print(f"Error: {e}")

# Correct way — create a new string
new_word = 'J' + word[1:]
print(new_word)  # 'Jython'
````

## 5. String Concatenation & Repetition

````python
# Concatenation with +
first = 'Hello'
last  = 'World'
print(first + ' ' + last)

# Repetition with *
print('ha' * 3)   # 'hahaha'
````

## 6. Common String Methods

````python
s = '  Hello, Python World!  '

print(s.strip())            # Remove leading/trailing whitespace
print(s.lower())            # Convert to lowercase
print(s.upper())            # Convert to uppercase
print(s.replace('Python', 'Beautiful'))  # Replace substring
print(s.strip().startswith('Hello'))     # Check start
print(s.strip().endswith('!'))           # Check end
print('Python' in s)                     # Membership check
````

### split() and splitlines()

````python
# split() breaks a string into a list by a delimiter
print('1,2,3'.split(','))       # ['1', '2', '3']
print('1:2:3'.split(':'))       # ['1', '2', '3']
print('hello world'.split())    # ['hello', 'world'] — space is default

# splitlines() breaks on newline characters
text = 'line1\nline2\nline3'
print(text.splitlines())        # ['line1', 'line2', 'line3']
````

### String Formatting

````python
name = 'Alice'
age  = 30

# f-string (recommended, Python 3.6+)
print(f'Name: {name}, Age: {age}')

# .format() method
print('Name: {}, Age: {}'.format(name, age))

# Positional indexing in format
print('Age: {1}, Name: {0}'.format(name, age))
````

### Slicing defaults cheat sheet

```
           step POSITIVE (+)         step NEGATIVE (-)
           ────────────────────       ────────────────────
direction  left → right               right → left
start      0  (leftmost)              -1 (rightmost)
end        len(s) (past last)         -(len+1) (past first)
```
