# Chapter 2: Variables and Data Types

A variable is like a labelled tag attached to a value — the tag gives you a name to refer back to that value later, without having to write it out every time.

```python
# Creating some variables
name = "Madui Seth"
age = 25
height = 5
weight = 65
student = True
College = "DU"

print(name)
print(age)
print(height)
print(weight)
print(student)
print(College)
```

Python is smart - it figured out what type each variable is by looking at what we stored in it. Notice how we stored different types of information.

## Using Variables

```python
# Now let's use them together
print("Hello, my name is", name)
print("I am", age, "years old")
print("My height is", height, "feet")
print("My College is", College)
```

---

## The 5 Main Data Types

Python distinguishes between several kinds of values — here are the five you will encounter most often.

### Integers (int) - Whole Numbers

```python
count = 100
negative = -5
zero = 0

print(count, negative, zero)
print(type(count))
```

### Floats (float) - Numbers with Decimals

```python
price = 19.99
temperature = -2.5
pi = 3.14159

print(price, temperature, pi)
print(type(price))
```

### Strings (str) - Text

```python
city = "New York"
message = "Learning Python is fun!"
empty = ""

print(city)
print(message)
print(type(city))
```

### Booleans (bool) - True or False

```python
is_raining = True
is_snowing = False
is_sunny = True

print(is_raining, is_snowing, is_sunny)
print(type(is_raining))
```

### None - Representing Nothing

```python
nothing = None

print(nothing)
print(type(nothing))
```

---

## Checking the Type

```python
age = 30
score = 95.5
school = "Oxford"

print(type(age))
print(type(score))
print(type(school))
```

---

## Variables Can Change

```python
score = 50
print("Starting score:", score)

score = 75
print("Improved score:", score)

score = score + 10
print("Imroved Final score:", score)
```

That's why they're called variables - they can vary and change over time.
