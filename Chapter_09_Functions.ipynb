# Chapter 9: Functions

## What are Functions?

A **function** is a named piece of logic that you define once and invoke whenever you need it — letting you break a program into focused, manageable units instead of duplicating the same steps throughout your code.

---

## Creating Simple Functions

```python
# Define a function
def greet():
    print("Hello! Welcome to Python!")

# Call the function
greet()
greet()
greet()
```

---

## Functions with Parameters

```python
# Function with a parameter
def greet_person(name):
    print(f"Hello, {name}!")

greet_person("Alice")
greet_person("Bob")
greet_person("Charlie")
```

```python
# Function with multiple parameters
def add(a, b):
    print(f"{a} + {b} = {a + b}")

add(5, 3)
add(100, 50)
add(7, 8)
```

---

## Return Values

Functions can **return** values back to you

```python
# Function that returns a value
def multiply(x, y):
    result = x * y
    return result

# Use the returned value
answer = multiply(5, 4)
print(f"Result: {answer}")

print(f"6  7 = {multiply(6, 7)}")
```

```python
# Common return value patterns
def is_even(n):
    return n % 2 == 0

print(f"Is 4 even? {is_even(4)}")
print(f"Is 7 even? {is_even(7)}")
```

---

## Default Parameters

```python
# Function with default value
def greet_formal(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

greet_formal("Alice")  # Uses default
greet_formal("Bob", "Hi")  # Uses provided value
```

## Practical Examples

```python
# Calculate area of rectangle
def rect_area(length, width):
    return length * width

print(f"Area: {rect_area(5, 3)} square units")
print(f"Area: {rect_area(10, 8)} square units")
```

```python
# Convert temperature
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

print(f"0C = {celsius_to_fahrenheit(0)}F")
print(f"25C = {celsius_to_fahrenheit(25)}F")
print(f"100C = {celsius_to_fahrenheit(100)}F")
```

```python
# Check if someone is old enough
def can_vote(age):
    if age >= 18:
        return f"Yes, {age} year old can vote"
    else:
        return f"No, {age} year old cannot vote yet"

print(can_vote(16))
print(can_vote(21))
```

---

## Functions with Lists

```python
# Get sum of list
def sum_list(numbers):
    total = 0
    for num in numbers:
        total += num
    return total

print(sum_list([1, 2, 3, 4, 5]))
print(sum_list([10, 20, 30]))
```

```python
# Find average
def average(numbers):
    if len(numbers) == 0:
        return 0
    return sum_list(numbers) / len(numbers)

scores = [85, 90, 78, 92]
print(f"Average: {average(scores)}")
```
