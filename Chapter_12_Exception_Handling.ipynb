## What are Exceptions?

**Exceptions** are errors that occur during program execution. Handle them gracefully!

## Try-Except Block

**Try** to run code, and **except** if there's an error

```python
# Basic try-except
try:
    result = 10 / 2
    print(f"Result: {result}")
except ZeroDivisionError:
    print("Cannot divide by zero!")
```

```python
# This will trigger the except block
try:
    result = 10 / 0  # Error!
    print(f"Result: {result}")
except ZeroDivisionError:
    print("Cannot divide by zero!")
    result = 0

print(f"Final result: {result}")
```

---

## Catching Multiple Exceptions

```python
try:
    num = int("hello")  # ValueError
except ValueError:
    print("Invalid number!")
except TypeError:
    print("Type error occurred!")
```

```python
# Catch multiple errors with one except
try:
    num = int("abc")
except (ValueError, TypeError):
    print("Either ValueError or TypeError occurred!")
```

```python
# Catch any exception with Exception
try:
    value = 10 / 0
except Exception as e:
    print(f"An error occurred: {e}")
```

---

## Try-Except-Else

```python
# Code in 'else' runs if NO exception
try:
    num = int("42")
except ValueError:
    print("Invalid number!")
else:
    print(f"Success! Number is {num}")
    print(f"Double: {num * 2}")
```

---

## Try-Except-Finally

```python
# Code in 'finally' ALWAYS runs
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Error: Cannot divide by zero")
finally:
    print("Cleanup code runs here!")
    print("(This runs whether error occurred or not)")
```

## Practical Examples

```python
# Safe list access
def safe_access(lst, index):
    try:
        return lst[index]
except IndexError:
        print(f"Index {index} out of range")
        return None

numbers = [10, 20, 30]
print(safe_access(numbers, 1))  # Works
print(safe_access(numbers, 10))  # Out of range
```

```python
# Safe dictionary access
def get_value(dict, key):
    try:
        return dict[key]
except KeyError:
        return "Key not found!"

person = {"name": "Alice", "age": 25}
print(get_value(person, "name"))
print(get_value(person, "email"))
```

```python
# Safe integer conversion
def get_age():
    while True:
        try:
            age = int(input("Enter your age: "))
            if age < 0:
                print("Age cannot be negative")
                continue
            return age
except ValueError:
            print("Please enter a valid number")

# Uncomment the line below to test
# age = get_age()
# print(f"Your age is {age}")
```

## File Handling with Exceptions

```python
# Safe file reading
def read_file(filename):
    try:
        with open(filename, "r") as file:
            return file.read()
except FileNotFoundError:
        print(f"File {filename} not found!")
        return None
except IOError:
        print(f"Error reading file {filename}")
        return None

content = read_file("nonexistent.txt")
```
